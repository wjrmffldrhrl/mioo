```python
from operator import eq
import re
```

```python
## 중복되는 라인과 공백 라인 제거
info_file = open('info.txt','rt',encoding='UTF8')

Non_duplicate_info = ''
info_line = info_file.readlines()

count = 1
befor_line = ' '
for line in info_line:
    if eq(line,befor_line):## 중복검사
        pass
    elif len(line) != 1: # 중복되지 않고 공백 라인이 아닐때
        print(count,')',line)
        Non_duplicate_info += line
        befor_line = line
        count+=1
```

```python
info_file.close()
```

```python
## 타이틀 찾는 패턴 만들기
info_title = ['INFO','COLOR','FABRIC','SIZE','MODEL','DETAIL VIEW','DETAIL']
pattern=''
for title in info_title:
    pattern += title+'|'
pattern = pattern[:-1]


## 타이틀과 내용 딕션에 넣기
title_info = re.findall(pattern,Non_duplicate_info)

split_info = re.split(pattern,Non_duplicate_info)
split_info = split_info[1:]
info_dic = {}
for index,line in enumerate(split_info):
    print(title_info[index],')',line)
    title_split = {title_info[index]:line}
    info_dic.update(title_split)
    
print(info_dic['INFO'])
```
